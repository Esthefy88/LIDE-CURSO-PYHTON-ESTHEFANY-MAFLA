## Comparacion entre cuatro paises segun datos de futbooll

### Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Datos de los partidfos de futbol

```python
partidos=pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt', delimiter = "\t"
                    )
partidos.head(15)
```

```python
##Creacion tabla numero de goles por pais
```

```python
tabla1=partidos[['anio','anfitrion', 'estadio', 'ciudad', 'partido_orden', 'fecha', 'equipo_1', 'equipo_1_final']]
tabla1.head()
```

```python
tabla1.info()
```

```python
tabla2=partidos[['anio','anfitrion', 'estadio', 'ciudad', 'partido_orden', 'fecha', 'equipo_2', 'equipo_2_final']]
tabla2 = tabla2.rename(columns={'equipo_2': 'equipo_1'})
tabla2 = tabla2.rename(columns={'equipo_2_final': 'equipo_1_final'})
tabla2.head()
```

```python
tabla2.info()
```

### Graficos

```python
goles = pd.concat([tabla1, tabla2], ignore_index=True)
goles
```

```python
goles_pais = partidos.groupby(['equipo_1', 'anio'])['equipo_1_final'].sum().reset_index().sort_values(by='equipo_1_final', ascending=False)
goles_pais.head(10)
```

### GARFICO DE BARRAS EN CUADRICULAS DIFERENTES

```python
paises_seleccionados = ['Uruguay', 'Mexico', 'Brasil', 'Argentina']
df_filtrado = goles_pais[goles_pais['equipo_1'].isin(paises_seleccionados)]

# Crear subgráficos separados para cada país
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, pais in enumerate(paises_seleccionados):
    datos_pais = df_filtrado[df_filtrado['equipo_1'] == pais]
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    ax.bar(datos_pais['anio'], datos_pais['equipo_1_final'])
    ax.set_xlabel('Año')
    ax.set_ylabel('Goles')
    ax.set_title(pais)

plt.tight_layout()

# Mostrar el gráfico de barras separado por cuadrículas
plt.show()
```

En el grafico de puede apreciar que exisen años en los que los cuatro paises no clasifican al mundial por lo que se hacen evidentes vacios o saltos en los años. De la misma manera este grafico nos muestra los goles maximos de acuerdo al año del mundial, es asi que para el caso de Uruguay y Argentina maejan un total maximo por mundial de 14 y 17 goles respetivamente, mientras que Brasil tine un valor maximo de 20 goles en uno de los mundiales y muy por debajo se encuentra Mexico que ha logrado un maximo de 5 goles unicamente

### GRAFICO DE PUNTOS EN UNA MISMA CUADRICULA

```python
paises_seleccionados =['Uruguay', 'Mexico', 'Brasil', 'Argentina']


# Crear un gráfico de puntos con colores diferentes para cada país
plt.figure(figsize=(10, 6))
for i, pais in enumerate(paises_seleccionados):
    datos_pais = goles_pais[goles_pais['equipo_1'] == pais]
    plt.scatter(datos_pais['anio'], datos_pais['equipo_1_final'], label=pais)

plt.xlabel('Año')
plt.ylabel('Goles')
plt.title('Comparación de goles entre países')
plt.legend()

# Mostrar el gráfico de puntos
plt.show()
```

En el grafico de puntos se puede observar que, la nube de puntos de Mexico esta por debajo del resto de nubes de punto, ademas de forma general la participacion de los paises escogidos se presenta mas a partir del año 1960, mientras que para los años anteriores unicamente aparecen puntos pertenecientes a los paises de Uruguay y Brasil
